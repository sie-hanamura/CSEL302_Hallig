# **5. Model Training**

```python
X_train, X_test, y_train, y_test = train_test_split(X , Y , test_size = 0.2, random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)
```

# **6. Predictions and Evaluations**

```python
y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error (y_test, y_pred_test)
r_sq = r2_score(y_test, y_pred_test)

print(f"Mean Sqaured Error: {mse}")
print(f"R-Squared: {r_sq}")
```

# **7. Feauture Importance Analysis**

```python
prediction = model.predict(X_test)
plt.figure(figsize=(20,12),facecolor='lightgray')
plt.scatter(y_test, prediction, color = 'cadetblue', label= 'Predicted')
plt.plot(y_test, y_test, color = 'blue', linestyle ='-', label = 'actual', linewidth = 4)
plt.xlabel('Predict', fontweight='bold')
plt.ylabel('Actual', fontweight='bold')
plt.title('Linear Regression Prediction',fontsize=16,fontweight='extra bold')

plt.show()

print('Coefficient', model.coef_)
print('Intercept', model.intercept_)
```
